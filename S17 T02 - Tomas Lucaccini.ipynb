# EJERCICIO 3

```python
full_purchases_table = purchases.merge(stores,on='store_id').merge(products,on='product_id').merge(product_types,on='product_type_id').merge(clients,on='client_id')
```

### Creating binary gender column

```python
full_purchases_table['is_female'] = 0
full_purchases_table.loc[full_purchases_table['gender'] == 'Mujer', 'is_female'] = 1
full_purchases_table.loc[full_purchases_table['gender'] == 'Hombre', 'is_female'] = 0
```

```python
full_purchases_table
```

### Female gender proportion by Location and Store Category

```python
sns.factorplot('location', 'is_female', data=full_purchases_table, size=6, col='category')
```

### Amount of money spent & amount of items purchased by Store

```python
f, ax = plt.subplots(1, 2, figsize=[16, 6])
full_purchases_table.groupby('location').sum()['total'].plot(kind='bar', ax=ax[0])
full_purchases_table.groupby('location').count()['total'].plot(kind='bar', ax=ax[1], color='red')
plt.show()
```

### Purchases by Gender

```python
full_purchases_table.groupby('gender').count()['purchase_id'].plot(kind='pie', autopct="%.1f%%")
```

### Amount of Purchases by Location and Gender

```python
sns.relplot(x='size', y='purchase_id', data=full_purchases_table,
            kind='line', hue='gender', col='location', col_wrap=4)
```
