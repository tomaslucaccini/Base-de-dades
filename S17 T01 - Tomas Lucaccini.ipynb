### Creación de tabla Compras

```python
command7 = """CREATE TABLE IF NOT EXISTS
purchases(purchase_id INTEGER PRIMARY KEY, product_id INTEGER, store_id INTEGER, client_id INTEGER, employee_id INTEGER, quantity FLOAT, total FLOAT, FOREIGN KEY(product_id) REFERENCES products(product_id), FOREIGN KEY(store_id) REFERENCES stores(store_id), FOREIGN KEY(client_id) REFERENCES clients(client_id), FOREIGN KEY(employee_id) REFERENCES employees(employee_id))"""
```

```python
cursor.execute(command7)
```

```python
purchases_df = pd.read_csv('purchases.csv', sep=';')
```

```python
purchases_df
```

#### Importación de líneas en DF a tabla SQL

```python
purchases_df.to_sql('purchases', connection, if_exists='replace', index = False)
```

```python
cursor.execute('''  
SELECT * FROM purchases
          ''')
results = cursor.fetchall()

print(results)
```

## Prueba de Query: Extracción de valores de Compras para tiendas con Locación en Girona

```python
cursor.execute('''  
SELECT * FROM purchases as purchases LEFT JOIN stores as stores ON purchases.store_id = stores.store_id WHERE stores.location = 'Girona'
          ''')
results = cursor.fetchall()

print(results)
```

## Query: Total de ventas para tiendas de Barcelona

```python
cursor.execute('''  
SELECT SUM(total) FROM purchases as purchases LEFT JOIN stores as stores ON purchases.store_id = stores.store_id WHERE stores.location = 'Barcelona'
          ''')
results = cursor.fetchall()

print('Las ventas acumuladas para las tiendas de Barcelona son de ', results)
```
